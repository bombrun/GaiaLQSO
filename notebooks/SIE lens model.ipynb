A notebook to generate random lensed QSOs to illustrate proper motion magnification. With Gaia DR2 one might expect to have for the first time some proper motion in strong lensed QSOs.

If $dy$ is the proper motion of a source at $y$ one get the proper motion magnification in each image as $dx = A^{-1} dy$, where $A$ is the distortion matrix a the location of the image $x$.

This notebook is based on the SIE lens model. There are many other different lens models. Similar consideration might be applied to any lens models that defined image location and distortion matrix.

# SIE 

The SIE model was introduced in Kormann, Schneider & Bartelmann (1994) http://adsabs.harvard.edu/abs/1994A%26A...284..285K

## Lens equation
Let $D_L$ be the angular diameter lens distance and $D_S>D_L$ be the source distance from the observer. Consider the source and lens planes, 2 parallel planes at distance $D_{LS}$ (Note $D_{LS} \neq D_S-D_L$). Let $Y$ be the source position on the source plane and $X$ be the image position on the lens plane. Let $O$ be the center of the lens, and $O'$ its projection on the source plane. Let $X' = \frac{D_S}{D_L} X$ be the projection of $X$ on the source plane. The lens equation is simply the  vector relation $O'Y+YX'=O'X'$ on the source plane, for small deflection $YX'=D_{LS} \lambda(X)$, where $\lambda$ is the deflection angle, and the lens equation can be written as:
$$
Y = \frac{D_S}{D_L} X - D_{LS} \lambda(X)
$$
By introducing the normalised angular position $x=X/X_0$ and $y=Y/Y_0$ with $X_0$ the scaling parameter and $Y_0=D_S/D_L X_0$, one get the purely angular relation
$$
y = x - \alpha(x)
$$
with 
$$
\alpha(x) = \frac{1}{\pi} \int_{R^2} \kappa(x′) \frac{x − x′}{(x − x′)^2}  d^2x′ 
$$

## Gravitational lens model in normalised coordinates
Using $\kappa(x)$, the normalised surface mass density, the lens equation can be simplified as
$$
\alpha(x) = \nabla \psi(x)
$$
with 
$$
\psi(x) = \frac{1}{\pi} \int_{R^2} \kappa(x′) \ln{\left|x − x′\right|} \,  d^2x′
$$

### SIS model
$$\kappa =  \frac{1}{2\,x}$$
$$\alpha = \frac{x}{|x|}$$
$$ x = y - \frac{x}{|x|}$$

$$x=(r,\varphi)$$

$$ r \cos{\varphi} \, e_1 + r \sin{\varphi} \, e_2 =  (y_1- \cos{\varphi}) \, e_1 + (y_2- \sin{\varphi}) \, e_2 $$

$$ r = (y_1- \cos{\varphi})\cos{\varphi} + (y_2- \sin{\varphi})\sin{\varphi} $$


$$(y_1 + \cos{\varphi})\sin{\varphi} - (y_2 + \sin{\varphi})\cos{\varphi} = 0$$


### SIE model 
Section 3 in Kormann, Schneider & Bartelmann (1994).

$$X_0 = 4 \pi \frac{v^2}{c^2} \frac{D_d D_{ds}}{D_{s}} $$ where $D_d, D_s, D_ds $ are the angular diameter distance to the lens, the source and the source to the lens respectively.

Normalised surface mass density for SIE model
$$\kappa =  \frac{\sqrt{f}}{2 \sqrt{x_1^2+f^2 x_2^2}}$$
with $f\in ]0,1]$ and $(x_1,x_2)=(r\cos\varphi,r\sin\varphi)$ normalised Cartisian coordinates of the impact parameter $x$. 

$$\kappa(r,\varphi) = \frac{\sqrt{f}}{2\, r \sqrt{\cos^2\varphi+f^2 \sin^2\varphi}}$$
One get SIS $\kappa(r) = 1/2r$ for $f=1$. 

The associated normalised reflection angle
$$\alpha = \frac{\sqrt{f}}{\sqrt{1-f^2}} 
\left[
\text{arcsinh}\left(
\frac{\sqrt{1-f^2}}{f} \cos{\varphi}
\right) e_1 
+
\text{arcsin}\left(
\sqrt{1-f^2} \sin{\varphi}
\right) e_2 
\right]$$
Strong lens are obtained when $\kappa > 1$

The distortion matrix $dy = A \, dx$ is defined as linearizing $y = x - \alpha(x)$,
$$ A = \begin{bmatrix}
1-2\kappa \sin^2(\varphi) & \kappa \sin(2\varphi) \\
\kappa \sin(2\varphi) & 1-2\kappa \cos^2(\varphi)
\end{bmatrix}
$$
i.e. if $dy$ is the proper motion of a source at $y$ one get the proper motion magnification in the images as $dx = A^{-1} dy$.

The potential can be decomposed and reduced to an ODE
$$\psi(r,\varphi) =  r  \, \tilde{\psi} \, (\varphi) $$
where
$$\tilde{\psi} =  \frac{\sqrt{f}}{\sqrt{1-f^2}} 
\left[
\cos{\varphi} \,
\text{arcsinh}\left(
\frac{\sqrt{1-f^2}}{f} \cos{\varphi}
\right) 
+
\sin{\varphi} \,
\text{arcsin}\left(
\sqrt{1-f^2} \sin{\varphi}
\right) 
\right]
$$

The SIE lens equation is then composed of
$$ r = y_1 \cos{\varphi} + y_2 \sin{\varphi} + \tilde{\psi} $$  
where the angles  $\varphi$  are obtained by solving
$$ \left[ 
y_1 +  \frac{\sqrt{f}}{\sqrt{1-f^2}} 
\text{arcsinh}\left(
\frac{\sqrt{1-f^2}}{f} \cos{\varphi}
\right) 
\right] \sin{\varphi} -
\left[ 
y_2 +  \frac{\sqrt{f}}{\sqrt{1-f^2}} 
\text{arcsin}\left(
\sqrt{1-f^2} \sin{\varphi}
\right) 
\right] \cos{\varphi} 
= 0
$$
with $(y_1,y_2)$ is the position of the source, and $(r \cos \varphi, r \sin \varphi)$ the position of the images.

illustration of the eigenvectors

```python
ax = plt.subplot(111,aspect='equal')
for phi,x in zip(phis,xs) :
    w,v = np.linalg.eig(np.linalg.inv(sie.A(x,phi,f)))
    x0 = x*np.cos(phi)
    x1 = x*np.sin(phi)
    dx0 = w[0]*v[0][0]/10
    dx1 = w[0]*v[1][0]/10
    #print(x0,x1,w[0],dx0,dx1)
    ax.arrow(x0,x1,dx0,dx1,color='C0',length_includes_head=True,shape='left')
    dx0 = w[1]*v[0][1]/10
    dx1 = w[1]*v[1][1]/10
    #print(x0,x1,w[1],dx0,dx1)
    ax.arrow(x0,x1,dx0,dx1,color='C0',length_includes_head=True,shape='left')
ax.scatter(xs*np.cos(phis),xs*np.sin(phis),s = np.exp(sie.magnification(xs,phis,f)))
ax.scatter(y1,y2,s=np.exp(1),label="source",color='C1')
plt.grid()
```
